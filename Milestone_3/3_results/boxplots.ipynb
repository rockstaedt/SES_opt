```python
##############
### TASK 2 ###
##############

vtb = []
for i in [10,100,1000,10000]:
    data = pd.read_csv(f"stochastic\\task_2\\{i}\\testing\\objective_values_4_1000.csv",header = None)
    x = data.values
    vtb.append(x[0])


# Create a figure instance
fig = plt.figure(1, figsize=(12, 7))

# Create an axes instance
ax = fig.add_subplot(111)

# Create the boxplot
bp = ax.boxplot(vtb)

## Custom x-axis labels
ax.set_xticklabels(['10', '100', '1000', '10000'])
ax.tick_params(axis='both', which='major', labelsize=14)
#labels
ax.set_xlabel('Sample Size',fontsize=18)
ax.set_ylabel('Objective Value',fontsize=18)
ax.set_title('Effect of Sample Size on Objective Value Task 2',fontsize=18)

# mark the mean    
means = [np.mean(x) for x in vtb]
pylab.scatter([1, 2, 3, 4], means)
```

```python
#### Mean & Variance Graph ###

### Task 1 ###
means = []
varis = []

for i in [10,100,1000,10000,100000]:
    data_mv = pd.read_csv(f"stochastic\\task_1\\{i}\\testing\\mean_var_4_1000.csv",header = None)
    x = data_mv.values
    means.append(float(x[1][0]))
    varis.append(float(x[1][1]))

sz = np.array([10,100,1000,10000,100000])
b = np.arange(1,6)
means = np.array(means)
varis = np.sqrt(np.array(varis))
yplus = means+varis
yminus = means-varis



fig = plt.figure(1, figsize=(12, 7))

plt.plot(b,means)
plt.xticks(b,sz)
plt.fill_between(b,yminus,yplus,alpha=0.2, edgecolor='#1B2ACC', facecolor='#089FFF')
plt.ylim([85,105])
plt.ylabel('Objective Value',fontsize=15)
plt.xlabel('Sample Size', fontsize=15)
plt.show()
```
