```python
import numpy as np
```

```python
LOADS = [0,8,8,10,10,10,16,22,24,26,32,30,28,22,18,16,16,20,24,28,34,38,30,22,12]
seed = 12
```

```python
def AV(values, iterations):
    all_samples = []
    
    if (iterations % 2) != 0:
        iterations +=1
        
        
    iterations = int(iterations/2)
    
    for n,j in enumerate(range(iterations)):
        one_sample = []
        mirror_sample =  []
        
        for m,i in  enumerate(values):
            np.random.seed(n+m)
            x = np.random.normal(i, i*(1/3), 1)[0]
            one_sample.append(x)
            
            
            xm = i -( x-i)
            mirror_sample.append(xm)
            
            
        all_samples.append(one_sample)
        all_samples.append(mirror_sample)

    return all_samples

results = AV(LOADS,4)
results
```
